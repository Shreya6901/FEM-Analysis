```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset
file_path = "FEM4_data.csv"  
df = pd.read_csv(file_path)
```

```python
# Initial coordinates
nodes = df[['X', 'Y']].values  
# Displacements
displacements = df[['u_x', 'u_y']].values  
# Final positions after displacement
final_positions = df[['X1', 'Y1']].values  
 # Stress values
stresses = df[['sigma_xx', 'sigma_yy', 'sigma_xy']].values 
 # Failure status
status = df['Status'].values 
```

```python
# Map failure status to numerical values (1 = Failure, 0 = Elastic)
failures_per_node = np.where(status == "Failure", 1, 0)
```

```python
# Plot 1: Displacement Field (Quiver Plot)
plt.figure(figsize=(10, 6))
plt.quiver(nodes[:, 0], nodes[:, 1], displacements[:, 0], displacements[:, 1],
           angles='xy', scale_units='xy', scale=0.1, color='blue', alpha=0.5)
plt.scatter(nodes[:, 0], nodes[:, 1], c=failures_per_node, cmap="coolwarm", marker='o')
plt.colorbar(label="Failure Condition (1=Failed, 0=Stable)")
plt.xlabel("X Position (m)")
plt.ylabel("Y Position (m)")
plt.title("Displacement Field and Failure Zones in Mine Slope")
plt.show()
```

```python
# Plot 2: Stress Distribution (Sigma_xx)
plt.figure(figsize=(10, 6))
plt.scatter(nodes[:, 0], nodes[:, 1], c=stresses[:, 0], cmap="coolwarm", marker='o')
plt.colorbar(label="Sigma_xx (Pa)")
plt.xlabel("X Position (m)")
plt.ylabel("Y Position (m)")
plt.title("Stress Distribution (Sigma_xx) in Mine Slope")
plt.show()
```

```python
# Plot 3: Stress Distribution (Sigma_yy)
plt.figure(figsize=(10, 6))
plt.scatter(nodes[:, 0], nodes[:, 1], c=stresses[:, 1], cmap="coolwarm", marker='o')
plt.colorbar(label="Sigma_yy (Pa)")
plt.xlabel("X Position (m)")
plt.ylabel("Y Position (m)")
plt.title("Stress Distribution (Sigma_yy) in Mine Slope")
plt.show()
```

```python
# Plot 4: Mohr-Coulomb Failure Zones based on dataset
plt.figure(figsize=(10, 6))
plt.scatter(nodes[:, 0], nodes[:, 1], c=failures_per_node, cmap="coolwarm", marker='o')
plt.colorbar(label="Failure Condition (1=Failed, 0=Stable)")
plt.xlabel("X Position (m)")
plt.ylabel("Y Position (m)")
plt.title("Mine Bench Slope - Mohr-Coulomb Failure Zones")
plt.show()
```
